# Importações

```python
# reads XML files 
import xml.etree.ElementTree as ET 
tree = ET.parse('datasets/ObterDeputados.xml') # api v1 this list is volatile and has 513 deputados
deputados = tree.getroot()

# deputados registered so far
import pandas as pd
df_registered = pd.read_csv('resource_uri/person_resource_uri.csv', sep=';', index_col=None)
print('deputados registered so far', df_registered.shape[0])

# converts xml 2 dataframe
from models.xml2dataframe import XML2DataFrame
```

 # Deputados em exercício

```python
xmlstr = ET.tostring(deputados, encoding='utf-8', method='xml')
xml2df = XML2DataFrame(xmlstr)
df = xml2df.process_data()

num_titulares = sum(df.loc[:,'condicao'] == 'Titular')
num_suplentes = sum(df.loc[:,'condicao'] == 'Suplente')
print('numero de deputados:',df.shape[0])
print(
      'titulares:{:}\tsuplentes:{:}'.format(
          num_titulares,
          num_suplentes)
)
    
df.head()
```
